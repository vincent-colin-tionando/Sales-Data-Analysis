# 4. Data Cleaning

```python
# Dropping the Row ID Column
df.drop('Row ID', axis=1, inplace=True)

# Converting the date column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# Converting the 'Postal Code' column to numeric format
df['Postal Code'] = df['Postal Code'].astype('Int64')
```

```python
df.info()
```

```python
# Sorting the dataset by order date
df.sort_values(by=['Order Date'], inplace=True, ascending=True)
```

```python
# We can see that the postal code is not mentioned only for Burlington city in Vermont state. So, we need to fill the postal code of that city.
df['Postal Code'] = df['Postal Code'].fillna(5401)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Displaying the duplicated rows in the dataset
df[df.duplicated()]
```

```python
# Checking whether there are any duplicate order IDs in the dataset
print(df[df['Order ID'] == 'US-2015-150119'].shape)
df[df['Order ID'] == 'US-2015-150119']
```

```python
# We should drop one of the duplicated rows to avoid any bias in our analysis.
df[(df['Product ID'] == 'FUR-CH-10002965') & (df['Customer ID'] == 'LB-16795')]
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Converting the 'Postal Code' column to string format
df['Postal Code'] = df['Postal Code'].astype(str)
```
